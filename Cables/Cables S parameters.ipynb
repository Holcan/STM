```python
import numpy as np 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
header_list = ["Frequency [Hz]","S11 Magnitude[db]","S11 Phase[degrees]","S21 Magnitude[db]","S21 Phase[degrees]", "S12 Magnitude[db]", "S12 Phase[degrees]", "S22 Magnitude[db]", "S22 Phase[degrees]"]
```

```python
kabel1 = pd.read_csv("Rotes Kabel 1 text.txt",names = header_list ,delimiter =  " ")
```

```python
kabel1
```

S11 and s22 should be symmetric, as well as s12,s21 plot them

```python
kabel1["Frequency [Hz]"][2500:]
```

```python
fig,  ax = plt.subplots()

ax.grid(True)

plt.plot(kabel1["Frequency [Hz]"][0:1000]/1e9,kabel1["S11 Magnitude[db]"][0:1000],label = "S11 parameter", color="red")
plt.plot(kabel1["Frequency [Hz]"][0:1000]/1e9,kabel1["S22 Magnitude[db]"][0:1000],label = "S22 parameter", color="blue")

ax.set_xlabel('Frequency [GHz]')
ax.set_ylabel('dB')
ax.set_title('Red Cable S Parameters Magnitude')
ax.legend(loc=0)
plt.show()

fig.savefig(r'S Parameters, red cable till 4GHz.jpeg',bbox_inches='tight')
```

```python
fig,  ax = plt.subplots()

ax.grid(True)

plt.plot(kabel1["Frequency [Hz]"][1000:2400]/1e9,kabel1["S11 Magnitude[db]"][1000:2400],label = "S11 parameter", color="red")
plt.plot(kabel1["Frequency [Hz]"][1000:2400]/1e9,kabel1["S22 Magnitude[db]"][1000:2400],label = "S22 parameter", color="blue")

ax.set_xlabel('Frequency [GHz]')
ax.set_ylabel('dB')
ax.set_title('Red Cable S parameters')
ax.legend(loc=0)
plt.show()

fig.savefig(r'S Parameters, red cable 4GHz-8GHz.jpeg',bbox_inches='tight')
```
