```python
import numpy as np 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
header_list = ["Frequency [Hz]","S11 Magnitude[db]","S11 Phase[degrees]","S21 Magnitude[db]","S21 Phase[degrees]", "S12 Magnitude[db]", "S12 Phase[degrees]", "S22 Magnitude[db]", "S22 Phase[degrees]"]
```

```python
kabel1 = pd.read_csv("Rotes Kabel 1 text.txt",names = header_list ,delimiter =  " ")
```

```python
fig,  ax = plt.subplots()

ax.grid(True)

plt.plot(kabel1["Frequency [Hz]"][1000:2500]/1e9,kabel1["S11 Magnitude[db]"][1000:2500],label = "S parameter Rotes Kabel 1 file", color="red")

ax.set_xlabel('Frequency [GHz]')
ax.set_ylabel('S11 dB')
ax.set_title('Red Cable S11 Magnitude')
plt.show()

#fig.savefig(r'S11 Parameter, red cable.jpeg'.format,bbox_inches='tight')
```
