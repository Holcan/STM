```python
import numpy as np 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
header_list = ["Hz","S","dB","R","50", "S11", "S21", "S12", "S22"]
```

```python
kabel1 = pd.read_csv("Rotes Kabel 1 text.txt",names = header_list ,delimiter =  " ")
```

```python
kabel1
```

```python
fig,  ax = plt.subplots()

ax.grid(True)

plt.plot(kabel1["Hz"],kabel1["S21"],label = "S parameter Rotes Kabel 1 file", color="red")

ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('S')
#ax.set_title('a')
plt.show()
```
