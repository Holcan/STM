```python
import numpy as np 
import matplotlib
import matplotlib.pyplot as plt 
from numpy import loadtxt
import pandas as pd
```

# 100 $\mu$

```python
data_100u = loadtxt('averaged signal 100 u.csv',delimiter=',')
timedelay_100u = np.linspace(-200,200,31)
```

```python
#chopping data
points_tofi_100u = np.concatenate((data_100u[:5], data_100u[21:]))
time_delay_tofit_100u = np.concatenate((timedelay_100u[:5],timedelay_100u[21:] ))

#fitting slop
fitted_slope_100u,fitted_intercep_100u = np.polyfit(time_delay_tofit_100u,points_tofi_100u,1)

#calculating autocorwith fit
fitted_autocor_100u = np.array([fitted_intercep_100u + fitted_slope_100u*x for x in timedelay_100u])

#normalizing
normalized_autocor_100u = data_100u/fitted_autocor_100u
```

```python
fitted_slope_100u,fitted_intercep_100u
```

```python
fig,axes = plt.subplots()

axes.grid(True)

plt.plot(timedelay_100u,normalized_autocor_100u,'-o',label=r'100$\mu$s Pulse Width ',color='green')
#plt.plot(tm,pul,label = '{a} samples'.format(a = len(pul)))

axes.set_xlabel(r'time delay [$\mu$s]')
axes.set_ylabel(r'STM Signal [arb. units]')
axes.set_title('Auto-Correlation 31 steps ')

axes.legend(loc = 0)
plt.show()

fig.savefig(r'Normalized Sigma Autocorrelation Function 100u, no shielding, 31 steps.jpeg')
```

```python
#chopping data
points_tofi_500nano = np.concatenate((data_500nano[:8], data_500nano[-8:]))
time_delay_tofit_500nano = np.concatenate((timedelay_500nano[:8],timedelay_500nano[-8:] ))

#fitting slop
fitted_slope_500nano,fitted_intercep_500nano = np.polyfit(time_delay_tofit_500nano,points_tofi_500nano,1)

#calculating autocorwith fit
fitted_autocor_500nano = np.array([fitted_intercep_500nano + fitted_slope_500nano*x for x in timedelay_500nano])

#normalizing
normalized_autocor_500nano = data_500nano/fitted_autocor_500nano
```
