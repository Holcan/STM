### Pulse function and Sweeping:

```python
def Pl(Pulse,t):
    """Primitive rectangular pulse, A gives the amplitude, t should be evaluated over the array of time"""
    """given by the points starting at Start time, through the duration given by Start Duration """
    
    pul= Pulse['Amplitude']* (  Pulse['Start time'] <=  t <= Pulse['Start Duration'] + Pulse['Start time'] )
   
    return pul
```

```python
def genSweep(Pulse1,Pulse2,time,dt):
    "Generates the two pulses Pulse 1 and Pulse 2, Pulse2 will be swept"
    "in the interval [0, Endtime] with a time step given by dt, b corresponds to the starting "
    "position of Pulse2"
    plu1=np.array([Pl(Pulse1,t) for t in time])
    for i in range(Pulse2['Start time'], Pulse2['End time']+dt,dt):
        plu2=np.array([ Pl(Pulse2,(t-i)) for t in time])
        pulss= np.add(plu1,plu2)
    return pulss
        
    
    
```

```python
def genPulse(Pulse1,Pulse2,t):
    "generates two pulses give their parameters at different times t in the interval tim"
    tims=np.linspace(-1,t,1000)
    plu1=np.array([Pl(Pulse1,i) for i in tims])
    plu2=np.array([Pl(Pulse2,i) for i in tims])
    plu3= np.add(plu1,plu2)
    return plu3,tims
```
